```python
import matplotlib.pyplot as plt

def plot_outliers(df):
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

    for col in numeric_cols:
        plt.figure(figsize=(6, 4))
        plt.boxplot(df[col].dropna(), vert=False)
        plt.title(f"Outliers in {col}")
        plt.xlabel(col)
        plt.show()

plot_outliers(df_encoded)
```

```python
def remove_outliers_iqr(df):
    df_clean = df.copy()
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]

    return df_clean

# Example usage
df_encoded= remove_outliers_iqr(df_encoded)
```

```python
def clip_outliers(df, lower_quantile=0.01, upper_quantile=0.99):
    df_clipped = df.copy()
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

    for col in numeric_cols:
        lower = df[col].quantile(lower_quantile)
        upper = df[col].quantile(upper_quantile)
        df_clipped[col] = df[col].clip(lower, upper)

    return df_clipped


df_encoded = clip_outliers(df_encoded, 0.01, 0.99)
```

```python
rng = np.random.default_rng(12)
def inject_noise(col: pd.Series, gauss_scale=0.5, outlier_frac=0.01):
    col = col.astype(float).copy()
    std = np.nanstd(col)
    if not np.isfinite(std) or std == 0:
        std = 1.0
    noise = rng.normal(0.0, std * gauss_scale, size=len(col))
    noisy = col + noise


    n_out = max(1, int(outlier_frac * len(noisy)))
    idx = rng.choice(len(noisy), size=n_out, replace=False)
    extreme_val = np.nanmedian(noisy) + 10 * std
    noisy.iloc[idx] = extreme_val
    return noisy

noisy_data['age'] = inject_noise(noisy_data['age'])
noisy_data['campaign'] = inject_noise(noisy_data['campaign'])
```

```python
def plot_hist_overlay(before: pd.Series, after: pd.Series, title: str, xlabel: str):
    plt.figure()
    all_vals = pd.concat([before, after], axis=0)
    bins = np.histogram_bin_edges(all_vals.dropna(), bins=30)
    plt.hist(before.dropna(), bins=bins, alpha=0.6, label="Before")
    plt.hist(after.dropna(), bins=bins, alpha=0.6, label="After")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Count")
    plt.legend()
    plt.show()

def plot_boxpair(before: pd.Series, after: pd.Series, title: str, ylabel: str):
    plt.figure()
    plt.boxplot([before.dropna(), after.dropna()],
                labels=["Before", "After"], showfliers=True)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.show()

# --- Plots ---
plot_hist_overlay(data['age'], noisy_data['age'], "age: Boxplot", 'age')
plot_boxpair(data['age'], noisy_data['age'], "age: Boxplot",  'age')

plot_hist_overlay(data['campaign'], noisy_data['campaign'], "campaign: Histogram",'campaign')
plot_boxpair(data['campaign'], noisy_data['campaign'], "'campaign': Boxplot", 'campaign')
```

```python
import numpy as np
import pandas as pd
from scipy.stats import zscore, mstats

def denoise(series: pd.Series, z_thresh: float = 3.0, winsor_limits=(0.01, 0.01)):
    s = series.copy().astype(float)

    z_scores = zscore(s, nan_policy='omit')
    s = s.mask(np.abs(z_scores) > z_thresh, np.nan)

    s = s.fillna(s.median())


    s = mstats.winsorize(s, limits=winsor_limits)

    return pd.Series(s, index=series.index)


noisy_data['age_denoised'] = denoise(noisy_data['age'])
noisy_data['campaign_denoised'] = denoise(noisy_data['campaign'])

plot_hist_overlay(data['age'], noisy_data['age_denoised'], "age: Boxplot", 'age')
plot_boxpair(data['age'], noisy_data['age_denoised'], "age: Boxplot",  'age')

plot_hist_overlay(data['campaign'], noisy_data['campaign_denoised'], "campaign: Histogram",'campaign')
plot_boxpair(data['campaign'], noisy_data['campaign_denoised'], "'campaign': Boxplot", 'campaign')
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PowerTransformer
import matplotlib.pyplot as plt

def reduce_skewness(df: pd.DataFrame, skew_thresh=0.75, clip_percentile=0.01):
    """
    Reduce skewness of numeric features using Yeo-Johnson transform safely.

    Steps:
    - Identify skewed numeric features
    - Clip extreme outliers to avoid numerical issues
    - Apply Yeo-Johnson transform safely
    - Return less-skewed DataFrame and skewness comparison
    """
    df = df.copy()
    numeric_cols = df.select_dtypes(include='number').columns
    numeric_cols = [col for col in numeric_cols if df[col].nunique() > 1]  # skip constants

    # Compute skewness
    skewness = df[numeric_cols].skew()
    skewed_cols = skewness[abs(skewness) > skew_thresh].index.tolist()
    print(f"Skewed features (skew>{skew_thresh}): {skewed_cols}")

    # Clip extreme outliers
    df_clipped = df.copy()
    for col in skewed_cols:
        lower = df[col].quantile(clip_percentile)
        upper = df[col].quantile(1 - clip_percentile)
        df_clipped[col] = df[col].clip(lower, upper)

    # Yeo-Johnson transform safely
    df_yj = df_clipped.copy()
    successful_cols = []
    for col in skewed_cols:
        try:
            pt = PowerTransformer(method='yeo-johnson', standardize=True)
            df_yj[col] = pt.fit_transform(df_yj[[col]])
            successful_cols.append(col)
        except Exception as e:
            print(f"Skipping {col} for Yeo-Johnson due to: {e}")

    # Skewness comparison
    skew_comparison = pd.DataFrame({
        'Original': df[skewed_cols].skew(),
        'Yeo-Johnson': df_yj[skewed_cols].skew()
    })

    # Plot comparison
    skew_comparison.plot(kind='bar', figsize=(12,6), edgecolor='black', color=['skyblue', 'orange'])
    plt.axhline(y=0, color='gray', linestyle='--')
    plt.title("Skewness: Original vs Yeo-Johnson")
    plt.ylabel("Skewness")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    print(f"Successfully transformed features: {successful_cols}")

    return df_yj, skew_comparison

# Example usage:
df_less_skewed, skew_comparison = reduce_skewness(data)
```

```python

# PCA + KMeans on numeric subset (for simplicity)
num_data = df_less_skewed_bool[num_cols].fillna(df_less_skewed_bool[num_cols].median())
scaled = StandardScaler().fit_transform(num_data)

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(scaled)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_pca)

sil = silhouette_score(X_pca, clusters)
print("Silhouette Score:", sil)

plt.scatter(X_pca[:,0], X_pca[:,1], c=clusters, cmap='viridis', alpha=0.6)
plt.title("PCA + KMeans Clusters")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python

wcss = []           # inertia (within-cluster sum of squares)
silhouette = []     # silhouette scores
K_range = range(2, 11)  # test k=2 to k=10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(scaled)
    wcss.append(kmeans.inertia_)
    sil_score = silhouette_score(scaled, clusters)
    silhouette.append(sil_score)

# Plot Elbow Curve and Silhouette Scores
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(K_range, wcss, marker='o')
plt.title("Elbow Method (WCSS vs k)")
plt.xlabel("Number of clusters (k)")
plt.ylabel("WCSS (Inertia)")

plt.subplot(1,2,2)
plt.plot(K_range, silhouette, marker='o')
plt.title("Silhouette Score vs k")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")

plt.tight_layout()
plt.show()
```
